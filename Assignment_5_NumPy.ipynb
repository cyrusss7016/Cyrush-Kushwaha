```
import numpy as np

gfg = np.matrix('4, 1, 9; 12, 3, 1; 4, 5, 6')

total_sum = gfg.sum()

row_sum = gfg.sum(axis=1)

col_sum = gfg.sum(axis=0)

print("Sum of all elements:", total_sum)
print("Sum of all elements row-wise:", row_sum)
print("Sum of all elements column-wise:", col_sum)
```

```
array = np.array([10, 52, 62, 16, 16, 54, 453])

sorted_array = np.sort(array)
print("Sorted array:", sorted_array)

sorted_indices = np.argsort(array)
print("Indices of sorted array:", sorted_indices)

smallest_elements = np.sort(array)[:4]
print("4 smallest elements:", smallest_elements)

largest_elements = np.sort(array)[-5:]
print("5 largest elements:", largest_elements)
```

```
import numpy as np

array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

integer_elements = array[array == array.astype(int)]

float_elements = array[array != array.astype(int)]

print("Integer elements only:", integer_elements)
print("Float elements only:", float_elements)
```

```
import numpy as np

initials = ['A', 'B']
X = sum(ord(char) for char in initials)

sales = np.array([X, X+50, X+100, X+150, X+200])
print("Sales dataset:", sales)

tax_rate = ((X % 5) + 5) / 100
taxed_sales = sales * (1 + tax_rate)
print("Sales after applying tax:", taxed_sales)

discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)
print("Sales after applying discount:", discounted_sales)

weekly_sales = np.tile(sales, (3, 1))
week_multipliers = np.array([1.00, 1.02, 1.04]).reshape(3, 1)
adjusted_weekly_sales = weekly_sales * week_multipliers

print("Weekly sales matrix:\n", adjusted_weekly_sales)
```

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)

plt.figure(figsize=(10, 6))

plt.plot(x, y1, label=r'$y = x^2$', linestyle='dashed', color='r')
plt.plot(x, y2, label=r'$y = \sin(x)$', linestyle='solid', color='b')
plt.plot(x, y3, label=r'$y = e^x$', linestyle='dotted', color='g')
plt.plot(x, y4, label=r'$y = \log(|x| + 1)$', linestyle='dashdot', color='m')

plt.title("Function Plots using NumPy and Matplotlib", fontsize=14)
plt.xlabel("X values", fontsize=12)
plt.ylabel("Y values", fontsize=12)
plt.legend()
plt.grid(True)

plt.show()
```
