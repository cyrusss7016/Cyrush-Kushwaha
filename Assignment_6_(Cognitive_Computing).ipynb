```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")

M = float(input("Enter a value for M: "))

x = np.linspace(-10, 10, 400)
y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, label=r'$y = M \cdot x^2$', color='blue', linestyle='--')
plt.plot(x, y2, label=r'$y = M \cdot \sin(x)$', color='red', linestyle='-')

plt.legend()
plt.title(f'Plot of Functions for M = {M}')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.show()
```

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Art'],
    'Score': [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Subject', y='Score', data=df, palette='muted')

for i, score in enumerate(df['Score']):
    ax.text(i, score + 1, str(score), ha='center', fontsize=12, fontweight='bold')

plt.title('Student Scores by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

roll_number = 12345
np.random.seed(roll_number)

data = np.random.randn(50)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(np.cumsum(data), color='blue', linestyle='-', marker='o')
axes[0, 0].set_title('Cumulative Sum Plot')
axes[0, 0].set_xlabel('Index')
axes[0, 0].set_ylabel('Cumulative Sum')
axes[0, 0].grid(True)

axes[0, 1].scatter(range(50), data, color='red', alpha=0.7)
axes[0, 1].set_title('Scatter Plot with Random Noise')
axes[0, 1].set_xlabel('Index')
axes[0, 1].set_ylabel('Value')
axes[0, 1].grid(True)

fig.suptitle('2x2 Subplot Layout', fontsize=14)
plt.tight_layout()
plt.show()
```

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
df = pd.read_csv(url)

plt.figure(figsize=(8, 5))
sns.lineplot(x=df.index, y=df['total_profit'], marker='o', color='b')
plt.title('Total Profit Over Months')
plt.xlabel('Months')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sales_columns = df.columns[1:-1]
for col in sales_columns:
    sns.lineplot(x=df.index, y=df[col], label=col)
plt.title('Monthly Sales Data of Different Products')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
df.mean().plot(kind='bar', colormap='viridis')
plt.title('Average Sales and Profit of All Features')
plt.xlabel('Features')
plt.ylabel('Average Value')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.show()
```
